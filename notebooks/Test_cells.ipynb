```python
from pyneuroml import pynml

ifs = ['HH2_477127614_IF.dat','HH2_476686112_IF.dat','AllenHH_477127614_IF.dat','AllenHH_476686112_IF.dat','L23_Retuned_477127614_IF.dat']
ivs = [i.replace('IF','IV') for i in ifs]
xs = []
ys = []
labels = []
linestyles = []
markers = []

for if_ in ifs:
    
    data, indeces = pynml.reload_standard_dat_file(if_)
    xs.append(data['t'])
    ys.append(data[0])
    labels.append(if_[:-7])
    linestyles.append('-')
    markers.append('o')
    
    

ax = pynml.generate_plot(xs,                      
                         ys,     
                         "IF curves from %s"%ifs,             
                         xaxis = 'Current (pA)',  
                         yaxis = 'Firing rate (Hz)',
                         linestyles=linestyles,
                         markers=markers,
                         labels=labels)
xs = []
ys = []
labels = []
linestyles = []
markers = []

for iv in ivs:
    
    data, indeces = pynml.reload_standard_dat_file(iv)
    xs.append(data['t'])
    ys.append(data[0])
    labels.append(if_[:-7])
    linestyles.append('-')
    markers.append('o')

ax = pynml.generate_plot(xs,                      
                         ys,     
                         "IV curves from %s"%ifs,             
                         xaxis = 'Current (pA)',  
                         yaxis = 'Membrane potential (mV)',
                         linestyles=linestyles,
                         markers=markers,
                         labels=labels)
    
    
    
```
